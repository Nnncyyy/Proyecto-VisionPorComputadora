```python
import os
import sys

# 1. Asegurar ruta raíz del proyecto
carpeta_actual = os.getcwd()
if not os.path.exists(os.path.join(carpeta_actual, "src")):
    raiz_proyecto = os.path.dirname(carpeta_actual)
else:
    raiz_proyecto = carpeta_actual

if raiz_proyecto not in sys.path:
    sys.path.insert(0, raiz_proyecto)

# 2. Importar todas las funciones necesarias
from src.video_utils import get_video_info, iter_video_frames, extraer_frames

# 3. Configurar rutas absolutas con os.path
video_path = os.path.join(raiz_proyecto, "data", "raw", "videoInstrucciones.mov")
output_dir = os.path.join(raiz_proyecto, "data", "processed", "frames")

print("🎬 Buscando video en:", video_path)
print("📂 Guardando frames en:", output_dir)
print("-" * 50)

# 4. Probar get_video_info (Si el archivo no existe, saltará un mensaje claro gracias a Path)
try:
    info_video = get_video_info(video_path)
    print("📊 Información del Video Detectada:")
    print(f"   - Resolución: {info_video['width']}x{info_video['height']}")
    print(f"   - FPS: {info_video['fps']:.2f}")
    print(f"   - Total de Frames: {info_video['total_frames']}")
except Exception as e:
    print(f"❌ Error al leer la info del video: {e}")

print("-" * 50)

# 5. Probar iter_video_frames (Recorrer los primeros 5 en memoria)
print("📸 Iterando sobre los primeros 5 frames mediante OpenCV nativo:")
try:
    for frame_idx, frame in iter_video_frames(video_path, max_frames=5, stride=1):
        print(f"Frame Index: {frame_idx} -> Dimensiones: {frame.shape}")
except Exception as e:
    print(f"❌ Error al iterar frames: {e}")

print("-" * 50)

# 6. Probar extraer_frames (Guardar en disco los 3 frames vía Supervision)
print("💾 Guardando imágenes físicas en el disco...")

try:
    extraer_frames(video_path=video_path, output_dir=output_dir, num_frames=3, stride=50)
except Exception as e:
    print(f"❌ Error en la extracción por Supervision (Verifica si el video está dañado): {e}")
```
