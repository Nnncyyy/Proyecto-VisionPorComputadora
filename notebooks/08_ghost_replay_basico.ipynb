# M3-03/M3-04 — Ghost Replay


Este notebook genera una imagen de Ghost Replay a partir de `tracks_projected.csv` o `tracks_with_color.csv`.

```python
from pathlib import Path
import sys, cv2, pandas as pd
ROOT = Path.cwd().resolve().parent if Path.cwd().name == "notebooks" else Path.cwd().resolve()
sys.path.insert(0, str(ROOT))
from src.ghost_replay import draw_ghost_replay

tracks_path = ROOT / "outputs" / "metrics" / "tracks_with_color.csv"
tracks = pd.read_csv(tracks_path)
frame_id = int(tracks["frame"].max())
canvas = 255 * __import__('numpy').ones((1200, 1600, 3), dtype='uint8')
result = draw_ghost_replay(canvas, tracks, current_frame=frame_id, memory=30)
out = ROOT / "outputs" / "figures" / "ghost_replay_sample.jpg"
out.parent.mkdir(parents=True, exist_ok=True)
cv2.imwrite(str(out), result)
print(out)
```
